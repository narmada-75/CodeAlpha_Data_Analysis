```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset("titanic")

print(df.head())
print(df.info())
print(df.shape)
print(df.isnull().sum())
print(df.describe())

sns.countplot(x='survived', data=df)
plt.title("Survival Count")
plt.show()

sns.countplot(x='sex', data=df)
plt.title("Gender Distribution")
plt.show()

sns.countplot(x='sex', hue='survived', data=df)
plt.title("Gender vs Survival")
plt.show()

sns.countplot(x='pclass', hue='survived', data=df)
plt.title("Passenger Class vs Survival")
plt.show()

sns.histplot(df['age'], bins=30)
plt.title("Age Distribution")
plt.show()

df['age'] = df['age'].fillna(df['age'].mean())

sns.boxplot(x=df['fare'])
plt.title("Fare Distribution")
plt.show()

df['sex'] = df['sex'].map({'male': 0, 'female': 1})

numeric_data = df.select_dtypes(include=['int64', 'float64'])

sns.heatmap(numeric_data.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()

print("Female passengers survived more than male passengers.")
print("First class passengers had a better survival rate.")
print("Missing values were present in the Age column.")
print("Fare column contains some outliers.")
print("EDA helped identify patterns and trends in the dataset.")
```
